### Part 3: Functions

```python
#Write a function to convert temperature from kelvin to celsius

def kelvin_to_celsius(temp_kelvin):
    """convert temperature from kelvin to celsius"""
    return temp_kelvin - 273.15

kelvin_to_celsius(273.15)

#and celsius to kelvin

def celsius_to_kelvin(temp_celsius):
    """convert temperature from celsius to kelvin"""
    return temp_celsius + 273.15

celsius_to_kelvin(0)
```

```python
#Write a function to convert temperature to fahrenheit

def temp_to_fahrenheit(temp, unit="celsius"):
    """convert temperature from celsius or kelvin to fahrenheit."""
    if unit == "celsius":
        return (temp*(9/5)) + 32
    elif unit == "kelvin":
        c = kelvin_to_celsius(temp)
        return (c*(9/5))+32
    
#Include an optional keyword argument to specify whether the input is in celcius or kelvin. Call your previously defined functions if necessary.
```

```python
#Check that the outputs are sensible by trying a few examples
temp_to_fahrenheit(100,'celsius')
```

```python
temp_to_fahrenheit(273.15,'kelvin')
```

```python
#Now write a function that converts from fahrenheit 
#and uses a keyword argument to specify whether you want the output in celsius or kelvin

def temp_from_fahrenheit(tempF, output="celsius"):
    """convert temperature from fahrenheit to celsius or kelvin."""
    if output == "celsius":
        return (tempF-32)*(5/9)
    elif output == "kelvin":
        return (tempF-32)*(5/9)+273.15
```

```python
temp_from_fahrenheit(32,'celsius')
```

```python
temp_from_fahrenheit(32,'kelvin')
```

```python
#Write a function that takes two arguments (feet and inches) and returns height in meters
#verify that it gives sensible answers

def feet_inch_to_meters(feet,inches):
    height_in_inches = 12*feet + inches
    height_in_meters = height_in_inches*0.0254
    return height_in_meters

feet_inch_to_meters(6,0)
```

```python
#Write a function takes one argument (height in meters) and returns two arguments (feet and inches)

def meters_to_feet_inch(meters):
    meters_to_inches = meters/0.0254
    feet = int(meters_to_inches//12)
    inches = meters_to_inches % 12
    return feet,inches
```
