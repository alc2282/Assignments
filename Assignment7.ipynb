# 7) Create a filtered dataset that only has earthquakes of magnitude 4 or larger

```python
df_filtered = df[df.mag>=4]
df_filtered.head()
```

# 8) Using the filtered dataset (magnitude > 4), count the number of earthquakes in each country/state. Make a bar chart of this number for the top 5 locations with the most earthquakes
Location name on the x axis, Earthquake count on the y axis

```python
df_filtered.groupby('country').mag.count().nlargest(5).plot(kind='bar')
plt.xlabel('Location (Country/State)')
plt.ylabel('Earthquake Count')
plt.title('Top 5 Locations With the Most Earthquakes of Magnitude >= 4')
```

# 9) Make a histogram of the distribution of the Earthquake magnitudes

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.hist.html https://matplotlib.org/api/_as_gen/matplotlib.pyplot.hist.html

Do one subplot for the filtered and one for the unfiltered dataset. Use a Logarithmic scale. What sort of relationship do you see?

```python
fig, axes = plt.subplots(ncols=2,figsize=(16,4))

axes[0].hist(df.mag, log=True, bins=np.arange(-1,8,0.1))
axes[0].set_title('Histogram of All Earthquake Magnitudes')
axes[0].set_xlabel('Magnitude')
axes[0].set_ylabel('log(Count)')

axes[1].hist(df_filtered.mag, log=True, bins=np.arange(-1,8,0.1))
axes[1].set_title('Histogram of Earthquakes with Magnitudes >= 4')
axes[1].set_xlabel('Magnitude')
axes[1].set_ylabel('log(Count)')
```

The histogram of earthquakes with a magnitude of 4 or larger accounts for the right-hand section of the histogram of all earthquakes.

# 10) Visualize the locations of earthquakes by making a scatterplot of their latitude and longitude

Use a two-column subplot with both the filtered and unfiltered datasets. Color the points by magnitude. Make it pretty

What difference do you note between the filtered and unfiltered datasets?

```python
fig, axes = plt.subplots(ncols=2,figsize=(16,4))

sc1= axes[0].scatter(
    df.longitude,
    df.latitude,
    c=df.mag,
    s=1.5,
    vmin=-1,
    vmax=8
)

axes[0].set_title('Scatter Plot of Earthquakes of All Magnitudes')
axes[0].set_xlabel('Longitude')
axes[0].set_ylabel('Latitude')
plt.colorbar(sc1, ax=axes[0], label='Magnitude')

sc2= axes[1].scatter(
    df_filtered.longitude,
    df_filtered.latitude,
    c=df_filtered.mag,
    s=1.5,
    vmin=-1,
    vmax=8
)

axes[1].set_title('Scatter Plot of Earthquakes of Magnitudes >= 4')
axes[1].set_xlabel('Longitude')
axes[1].set_ylabel('Latitude')
plt.colorbar(sc1, ax=axes[1], label='Magnitude')
```

The scatterplot of all earthquake magnitudes includes a cluster of smaller earthquakes in North America that were not of magnitude 4 or greater.
